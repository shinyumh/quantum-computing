```python
import matplotlib.pyplot as plt
import numpy as np

# f1
x = np.linspace(-3 * np.pi, 3 * np.pi, 200)
y = np.sin(0.8 * x)
plt.plot(x, y, label = r"$y=\sin{0.8x}$")

# f2
y = np.sin(0.5 * x)
plt.plot(x, y, label = r"$y=\sin{0.5x}$")

# f3
y = np.sin(0.8 * x) * np.sin(0.5 * x)
plt.plot(x, y, label = r"$y=\sin{0.8x}*\sin{0.5x}$")

# f4 - werner formula: 2sin(a)sin(b) = cos(a-b) - cos(a+b)
x = np.linspace(-3 * np.pi, 3 * np.pi, 100)
y = (np.cos(0.3 * x) - np.cos(1.3 * x)) / 2
plt.plot(x, y, "o:",mfc='none',label = r"$y=\frac{\cos{0.3x}-\cos{1.3x}}{2}$")

# plot graphs
plt.xlim(-3 * np.pi,3 * np.pi)
plt.legend(loc="upper right")
plt.show()
```
