```python
"""complex_lattice.ipynb"""
# cell 1

from __future__ import annotations

import matplotlib.pyplot as plt
import numpy as np

%matplotlib widget

# the complex function given
def f(x):
    return x ** 2 + x * 1j + 1

# set the domain and the range
x = np.linspace(-4,4,200)
y = np.linspace(0,2,200)
g = f(x + y * 1j)

plt.scatter(np.real(g), np.imag(g))
plt.xlabel("real")
plt.ylabel("imaginary")
plt.show()
```

```python
def calc(x,y):
    """for counting |real(z)| <= x and |imag(z)| <= y"""
    return 2 * x * 2 * y

print("there are " + str(calc(10,10)) + " gaussian integers for |real(z)| <= 10 and |imag(z)| <= 10 \n")

def calc2(a,b):
    """for counting |real(f(z))| <= a and |imag(f(z))| <= b"""
    counter = 0

    # try every number in between -a and a
    for x in range(0 - a,a + 1):

        # try every number in between -b and b
        for y in range(0 - b,b + 1):

            # check to see if the abs values of the two exps are <= 10
            if abs(x ** 2 - y ** 2 - y + 1) <= 10 and abs(2 * x * y + x) <= 10:
                counter += 1

    return counter

print("there are " + str(calc2(10,10)) + " gaussian integers for |real(f(z))| <= 10 and |imag(f(z))| <= 10 \n")
```
